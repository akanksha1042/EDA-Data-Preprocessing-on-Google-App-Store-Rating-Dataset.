```python
#1. Import required libraries and read the dataset.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy.stats import norm
from sklearn.preprocessing import LabelEncoder

import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv(r'C:\Users\Akanksha\Downloads\honeyproduction (1).csv')
```

```python
#2. Check the first few samples, shape, info of the data and try to familiarize yourself with different features.

print('Shape of the dataset is :',df.shape)
print('\n\n',df.info())

df.head()
```

3. Display the percentage distribution of the data in each year using the pie chart.

```python
# Group the data by year and calculate the sum of each column for each year
yearly_data = df.groupby('year').sum()

# Plot the pie chart
yearly_data.plot.pie(y='totalprod', figsize=(10, 10), autopct='%1.1f%%')
plt.title('Percentage Distribution of Total Honey Production by Year')
plt.ylabel('totalprod')
plt.show()
```

4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
sns.displot(df['priceperlb'], kde=True)
plt.title('Distribution of Price per lb')
plt.xlabel('Price per lb')
plt.ylabel('Frequency')
plt.show()
```

5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

```python
plt.scatter(df['numcol'], df['prodvalue'])
plt.title('Number of Colonies vs. Production Value')
plt.xlabel('Number of Colonies')
plt.ylabel('Production Value ($)')
plt.show()
```

```python
"""Inference : We can see that the fields prodvalue and  numcol is strongly correlated with each other 
and as numcol increases, the prodvalue increases."""
corr_coef = df['numcol'].corr(df['prodvalue'])
print('Correlation Coefficient:', corr_coef)
```

6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.


```python
plt.figure(figsize=(10, 6)) # set figure size to 10x6 inches
sns.boxplot(x='year', y='prodvalue', data=df)
plt.title('Production Value by Year')
plt.xlabel('Year')
plt.ylabel('Production Value ($)')
plt.show()
```
